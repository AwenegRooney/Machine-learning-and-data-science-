## Classifying iris flowers

```python
#Importing the neccesary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, f1_score
```

```python
iris = load_iris()
iris.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, random_state=42)
```

```python
model = RandomForestClassifier(n_jobs=2, random_state=42)
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)
```

```python
cm_train = confusion_matrix(y_train, y_train_pred)
fig = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=iris.target_names)
fig.plot(cmap=plt.cm.Blues)
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)
fig = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=iris.target_names)
fig.plot(cmap=plt.cm.Blues)
```

```python
print("Test accuracy: {}".format(accuracy_score(y_test, y_test_pred)))
print("Train accuracy: {}".format(accuracy_score(y_train, y_train_pred)))
print("Test f1: {}".format(f1_score(y_test, y_test_pred, average='weighted')))
print("Train f1: {}".format(f1_score(y_train, y_train_pred, average='weighted')))
```
