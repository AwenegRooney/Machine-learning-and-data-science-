## Classifying emails as spam or ham

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix, f1_score
```

```python
df = pd.read_csv('../../../Data Science/DataSets/email_classification_dataset.csv')
df.head()
```

```python
#checking to see if there are any missing data
df.isnull().sum()
```

```python
#the class of the email is independent of the id
df.drop(columns='id', inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['email'].values, df['label'].values, test_size=0.25, random_state=42)
```

```python
model = make_pipeline(TfidfVectorizer(), MultinomialNB())
model.fit(X_train, y_train)
```

```python
y_test_pred = model.predict(X_test)
y_train_pred = model.predict(X_train)

```

```python
cm_train = confusion_matrix(y_train, y_train_pred)
fig = ConfusionMatrixDisplay(confusion_matrix=cm_train, display_labels=model.classes_)
fig.plot(cmap=plt.cm.Greens)
```

```python
cm_test = confusion_matrix(y_test, y_test_pred)
fig = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=model.classes_)
fig.plot(cmap=plt.cm.Greens)
```

```python
#Checking accuracy
print("Test accuracy: {}".format(accuracy_score(y_test, y_test_pred)))
print("Train accuracy: {}".format(accuracy_score(y_train, y_train_pred)))
```
