### Seeing the differences between K-means and DBSCAN

```python
#Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
```

```python
#creating the data
features, true_labels = make_moons(n_samples=250, noise=0.05, random_state=42)
```

```python
pd.DataFrame(features).head()
```

```python
pd.DataFrame(true_labels).head()
```

```python
#Visualizing the data
plt.scatter(features[:, 0], features[:, 1])
```

```python
colors = ['maroon', 'forestgreen']
vectorizer = np.vectorize(lambda x: colors[x])
```

```python
#Visualizing how the clusters should be
plt.scatter(features[:, 0], features[:, 1], c=vectorizer(true_labels))
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
pd.DataFrame(scaled_features).head()
```

```python
#Creating the model
k_model = KMeans(n_clusters=2)
d_model = DBSCAN(eps=0.3)

#training the models
k_model.fit(scaled_features)
d_model.fit(scaled_features)
```

```python
#visualizing how KMeans grouped the data
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=vectorizer(k_model.labels_))
```

```python
#visualizing how KMeans grouped the data
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=vectorizer(d_model.labels_))
```

From observation, DBSCAN was able to group the data better than KMeans
